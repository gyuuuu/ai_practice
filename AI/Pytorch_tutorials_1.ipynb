```
from __future__ import print_function
import torch
```

```
x = torch.empty(5,3) # 5 by 3 의 행렬, 초기화 x
print(x)
```

```
x = torch.rand(5,3) #무작위로 초기화된 행렬 생성
print(x)
```

```
x = torch.zeros(5,3, dtype=torch.long)
print(x)
```

```
x = torch.tensor([5.5, 3])
print(x)
```

```
x = x.new_ones(5,3, dtype=torch.double)
print(x)

x = torch.randn_like(x, dtype=torch.float)
print(x)
```

```
print(x.size())
```

```
y = torch.rand(5,3)
print(x+y)
```

```
print(torch.add(x, y))
```

```
# 덧셈결과를 인자로 제공
result = torch.empty(5,3)
torch.add(x, y, out=result)
print(result)
```

```
y.add_(x)
print(y)
# 바꿔치기 (in-place)방식으로 tensor값을 변경하는 연산뒤에는 _가 붙는다.
```

```
print(x)
print(x[:, 1])
```

```
# tensor의 크기변경
x = torch.randn(4,4)
y = x.view(16)
z = x.view(-1, 8) #-1은 알아서 맞춰줌-> 원래 tensor와 총 크기가 같도록
print(x.size(), y.size(), z.size())
```

```
# tensor에 하나의 값만 존재한다면, .item()을 사용하여 값을 얻을 수 있음
x = torch.randn(1)
print(x)
print(x.item())
```

```
a = torch.ones(5)
print(a)
```

```
b = a.numpy()
print(b)
```

```
a.add_(1)
print(a)
print(b)
# torch tensor와 numpy 배열은 메모리 공간을 공유하기때문에 하나를 변경하면 다른하나도 변경됨
```

```
import numpy as np
a = np.ones(5)
b = torch.from_numpy(a)
np.add(a, 1, out=a)
print(a)
print(b)
```
