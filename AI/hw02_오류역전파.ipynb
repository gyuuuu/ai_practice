```
import numpy as np
```

```
np.random.seed(0)

N, D = 3, 4

x = np.random.randn(N, D)
y = np.random.randn(N, D)
z = np.random.randn(N, D)
```

```
a = x * y
b = a + z
c = np.sum(b)
```

```
grad_c = 1.0
grad_b = grad_c * np.ones((N, D))

# sum 이니까 그래도 전파 fanout
grad_a = grad_b.copy()
grad_z = grad_b.copy()

# 곱이니까 다른 쪽이랑 곱해짐
grad_x = grad_a * y
grad_y = grad_a * x

print("grad_c : ", grad_c)
print("grad_b : ", grad_b)
print("grad_a : ", grad_a)
print("grad_z : ", grad_z)
print("grad_x : ", grad_x)
print("grad_y : ", grad_y)
```

```
import torch

x = torch.randn(N,D)
x.requires_grad_(True)
y = torch.randn(N,D)
y.requires_grad_(True)
z = torch.randn(N,D)
z.requires_grad_(True)

a = x * y
b = a + z
c = torch.sum(b)

c.backward()
print("x미분 : ", x.grad)
print("y미분 : ", y.grad)
```
